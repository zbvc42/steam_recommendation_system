```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams.update({'font.size': 14})
```

```python
list_x = []
list_y = []
list_c = []
for x in np.linspace(0, 1, 1000):
    for y in np.linspace(0, 1, 1000):
        list_x.append(x)
        list_y.append(y)
        list_c.append(x * y)

plt.hexbin(list_x, list_y, list_c, vmin=0, vmax=1)
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Mean of X and Y')
plt.title('Mean of Two Numbers')
plt.show()
```

```python
list_x = []
list_y = []
list_c = []
for x in np.linspace(0, 1, 1000):
    for y in np.linspace(0, 1, 1000):
        list_x.append(x)
        list_y.append(y)
        list_c.append(2/(x**-1 + y**-1))

plt.hexbin(list_x, list_y, list_c, vmin=0, vmax=1)
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='Harmonic Mean of X and Y')
plt.title('Harmonic Mean of Two Numbers')
plt.show()
```

```python
list_x = []
list_y = []
list_c = []
for x in np.logspace(1, 7, 1024):
    for y in np.logspace(0,-4,1024):
        list_x.append(x)
        list_y.append(-np.log10(y)*10/7)
        list_c.append((((1-y)**np.log10(x)*10/7)*(np.log10(np.log10(x)*10/7))))

plt.hexbin(list_x, list_y, list_c, xscale='log', cmap='inferno')
plt.xlim((10, 10**7))
plt.ylim((0, 4))
plt.yticks([0, 1, 2, 3, 4],["0%", "90%", "99%", "99.9%", "99.99%"])
plt.xlabel('Number of Reviews')
plt.ylabel('Positive Review Percent')
plt.colorbar(label='Popularity', ticks=[])
plt.title('Popularity Metric')
plt.show()
```
